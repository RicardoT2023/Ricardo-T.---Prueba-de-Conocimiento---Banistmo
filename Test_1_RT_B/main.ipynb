## question 1: Basic SQL
Write a SQL statement from table <i>train_table</i> to obtain the top 10 <i>locations</i> people twit from (in descending order)

```python
# write query inside docstring
query_1 = """
        SELECT location, COUNT(*) AS count
FROM train_table
WHERE location IS NOT NULL
GROUP BY location
ORDER BY count DESC
LIMIT 10;

"""
```

```python
import csv
from typing import List


class Data:
    def __init__(self, data: List[List[str]]) -> None:
        self.data = data

    def infer_dtypes(self) -> List[type]:
        num_columns = len(self.data[0])
        types = []
        for i in range(num_columns):
            col_types = set()
            for j in range(1, len(self.data)):
                value = self.data[j][i]
                if value == '':
                    continue
                if value.isdigit():
                    col_types.add(int)
                else:
                    try:
                        float(value.replace('%', ''))
                        col_types.add(float)
                    except ValueError:
                        col_types.add(str)
            if len(col_types) > 1:
                types.append(str)
            else:
                types.append(col_types.pop())
        return types

    def describe(self) -> List[dict]:
        types = self.infer_dtypes()
        num_columns = len(self.data[0])
        stats = []
        for i in range(num_columns):
            if types[i] in [int, float]:
                column = []
                for j in range(1, len(self.data)):
                    value = self.data[j][i]
                    if value == '':
                        column.append(float('NaN'))
                    elif value.isdigit():
                        column.append(int(value))
                    else:
                        try:
                            column.append(float(value.replace('%', '')))
                        except ValueError:
                            column.append(float('NaN'))
                mean = sum(column) / len(column)
                std = (sum([(x - mean)**2 for x in column]) /
                       (len(column) - 1))**0.5
                max_val = max(column)
                min_val = min(column)
                stats.append(
                    {'column': self.data[0][i], 'mean': mean, 'std': std, 'max': max_val, 'min': min_val})
        return stats


class DataReader:
    @staticmethod
    def read_csv(file_path: str) -> Data:
        with open('./data/malformed_dataset.csv', 'r') as f:
            reader = csv.reader(f)
            data = []
            for row in reader:
                data.append(row)
        return Data(data)


```
